# Reading the image.

Firstly, we import the Image package from Pillow Imaging Library and then we read the image from the system.

```python
import sys
from PIL import Image

img = Image.open('MyPhoto.jpeg')
```

Resizing the image so that it can fit our notebook.

```python
print("Initial size of the Image")
print(img.size)
```

```python
width, height = img.size
aspect_ratio = height/width
new_width = 100
new_height = aspect_ratio * new_width * 0.55
img = img.resize((new_width, int(new_height)))
print("New Width")
print(new_width)
print("New Height")
print(int(new_height))
print("New size of the Image")
print(img.size)
```

Converting image into grayscale format

```python
img = img.convert('L')
```

Getting hold of all the pixels into a variable.

```python
pixels = img.getdata()
```

Replacing all the pixels with characters

```python
chars = ['@','#', 'S', '%', '?', '*', '+', ';', ':', ',',  '.']
new_pixels = [chars[pixel//25] for pixel in pixels]
new_pixels = ''.join(new_pixels)
```

Splitting the string of chars to fit our notebook and making list out of it 

```python
new_pixels_count = len(new_pixels)
ascii_image = [new_pixels[index:index + new_width] for index in range(0, new_pixels_count, new_width)]
ascii_image = "\n".join(ascii_image)
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import csv
 
def make_label_dict(labels):
    l = {}
    for i, label in enumerate(labels):
        l[i] = label
    return l
 
input_data = pd.read_csv('arr_aayush.csv', index_col=0)
print ("INPUT DATA")
print("\t")
print(input_data.values)
print("\n \n")
print("\n \n")
G = nx.Graph(input_data.values)
 
with open('labels.csv', 'r') as f:
    d_reader = csv.DictReader(f)
    headers = d_reader.fieldnames

print("HEADERS")
print("\t")
print(headers)
print("\n \n")
print("\n \n")
 
label_arr= make_label_dict(headers)
 
edge_labels = dict( ((u, v), d["weight"]) for u, v, d in G.edges(data=True) )
pos = nx.spring_layout(G)
nx.draw(G, pos)
#nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
nx.draw(G,pos,node_size=100, labels=label_arr, with_labels=False)
plt.show()
```

```python
nx.write_gml(G, "Aayush_graph.gml")
```
